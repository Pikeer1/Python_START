```python
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol('x')
```

```python
f = x**3 - 50*x
g = -x**4 + 88*x**2 - 241
h = f - g
```

```python
k = solve(h)
[[xelement.n().round(2),f.subs(x,xelement).n().round(2)] for xelement in k]
```

```python
plot(f,g)
```

```python
plot(h)
```
